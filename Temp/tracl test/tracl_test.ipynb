```python
import traci
import sumolib
```

```python
# Path to SUMO executable and config file
sumoBinary = "sumo"  # use "sumo-gui" if you want GUI, use "sumo" if you don't want GUI
#sumoConfig = "../temp4/Test.sumocfg"
sumoConfig = "./env/3x3grid.sumocfg"
```

```python
traci.start([sumoBinary, "-S", "-Q", "-c", sumoConfig])
print(f"number of traffic lights: {traci.trafficlight.getIDCount()}")
traci.close()
```

```python
try:

    traci.start([sumoBinary, "-S", "-Q", "-c", sumoConfig, "--verbose"]) # -S will automatically start the simulation in gui mode, -Q will close the gui.
    
    for step in range(100):
        traci.simulationStep()
        print(f"Step {step}: {traci.vehicle.getIDCount()} vehicles")
finally:
    traci.close()
```
